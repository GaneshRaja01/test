```python
df_tesla=df_tesla.set_index('Date',drop=True)
df_tesla.head()
```

```python
### https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
```

```python
## Year and frequency(rule A)
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
## plotting
df_tesla['Open'].resample(rule='A').max().plot(kind='bar')
```

```python
#MonthEnd
df_tesla['Open'].resample(rule='M').max().plot(kind='bar', figsize=(12,4))
```

```python
## Mean Average(rolling) for 10days + Aggretate function
#rolling can actually do smothening
df_tesla['High'].rolling(10).mean()
```

```python
df_tesla
```

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(30).mean()
df_tesla.head(32)
```

```python
df_tesla[['Open','Open:30 days rolling']].plot(figsize=(12,4))
```
